```python
train_datagen = ImageDataGenerator(rescale = 1./255, shear_range = 0.2,zoom_range = 0.2,
horizontal_flip = True,brightness_range=[0.5,0.3],rotation_range=0.2)
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
X_train_Imagepath='C:/Users/user/Desktop/I neuron DLNLPCV/DL DATASETS/JanataHack Computer Vision Hackathon/X_Train'
X_test_Imagepath='C:/Users/user/Desktop/I neuron DLNLPCV/DL DATASETS/JanataHack Computer Vision Hackathon/X_test'
```

```python
training_set = train_datagen.flow_from_directory(X_train_Imagepath, target_size = (224, 224),batch_size = 32,
class_mode = 'categorical')
```

```python
test_set = test_datagen.flow_from_directory(X_test_Imagepath,target_size = (224, 224),
batch_size = 32,class_mode = 'categorical')
```

```python
classes=glob('C:/Users/user/Desktop/I neuron DLNLPCV/DL DATASETS/JanataHack Computer Vision Hackathon/X_Train/*')
```

```python
################################################################################################
```

```python
vgg = VGG16(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
for layer in vgg.layers:
    layer.trainable = False
```

```python
vgg.summary()
```

```python
x = Flatten()(vgg.output)
x=Dense(100, activation='relu')(x)
prediction = Dense(len(classes), activation='softmax')(x)
```

```python
model = Model(inputs=vgg.input, outputs=prediction)
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
MODEL = model.fit_generator(training_set,validation_data=test_set,epochs=3,steps_per_epoch=len(training_set),
validation_steps=len(test_set))
```
